```python
import matplotlib.pyplot as plt

# Datos de entrenamiento
frutas = {
    "peso": [130, 150, 170, 200, 180, 210, 160, 140, 250, 230, 100, 190, 220, 280, 270, 300, 320, 330, 340, 350],
    "textura": [0, 0, 0, 1, 1, 1, 0, 0, 1, 1, 0, 1, 1, 1, 0, 0, 0, 1, 1, 1],
    "tipo": ["🍏", "🍏", "🍏", "🍊", "🍊", "🍊", "🍏", "🍏", "🍊", "🍊",
             "🍏", "🍊", "🍊", "🍊", "🍉", "🍉", "🍉", "🍍", "🍍", "🍍"]
}

# Graficar los datos
plt.scatter(frutas["peso"][:7], frutas["textura"][:7], color="green", label="Manzana 🍏", s=100)
plt.scatter(frutas["peso"][7:14], frutas["textura"][7:14], color="orange", label="Naranja 🍊", s=100)
plt.scatter(frutas["peso"][14:17], frutas["textura"][14:17], color="red", label="Sandía 🍉", s=100)
plt.scatter(frutas["peso"][17:], frutas["textura"][17:], color="yellow", label="Piña 🍍", s=100)
plt.scatter(175, 1, color="black", marker="*", label="Fruta Misteriosa ❓", s=200)

plt.xlabel("Peso (g)")
plt.ylabel("Textura (0 = lisa, 1 = rugosa)")
plt.legend()
plt.title("Clasificación de Frutas con kNN")
plt.show()
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Datos de entrenamiento
X = np.array([
    [130, 0], [150, 0], [170, 0], [200, 1], [180, 1], [210, 1], [160, 0], [140, 0], [250, 1], [230, 1],
    [100, 0], [190, 1], [220, 1], [280, 1], [270, 0], [300, 0], [320, 0], [330, 1], [340, 1], [350, 1]
])  # Peso y textura

y = np.array(["🍏 Manzana", "🍏 Manzana", "🍏 Manzana", "🍊 Naranja", "🍊 Naranja", "🍊 Naranja",
              "🍏 Manzana", "🍏 Manzana", "🍊 Naranja", "🍊 Naranja",
              "🍏 Manzana", "🍊 Naranja", "🍊 Naranja", "🍊 Naranja",
              "🍉 Sandía", "🍉 Sandía", "🍉 Sandía", "🍍 Piña", "🍍 Piña", "🍍 Piña"])  # Clases

# Crear modelo kNN con k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)  # Entrenar el modelo

# Nueva fruta a clasificar
nueva_fruta = np.array([[300, 0]])  # Peso 175g, textura rugosa (1)
prediccion = knn.predict(nueva_fruta)

print(f"🔮 Predicción del modelo: {prediccion[0]}")
```
